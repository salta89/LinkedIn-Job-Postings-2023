# 2. Cleaning Data

```python
# Drop unnecessary columns
df.drop(columns = ['description',
                   'pay_period',
                   'original_listed_time',
                   'job_posting_url',
                   'application_url',
                   'skills_desc',
                  'posting_domain',
                  'sponsored',
                  'work_type', 
                   'currency',
                   'expiry',
                   'scraped',
                  'closed_time',
                  'remote_allowed',
                  'compensation_type'],inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
# Convert each time column to datetime format

df['listed_time'] = pd.to_datetime(df['listed_time'], unit='ms')  
```

```python
# Renaming columns 
df.rename(columns={'formatted_work_type': 'work_type'}, inplace=True)
df.rename(columns={'formatted_experience_level': 'experience_level'}, inplace=True)
```

```python
df.dtypes
```

```python
#Finding duplicates 
duplicate_rows = df[df.duplicated()]

# Display the duplicate rows
print("Duplicate Rows except first occurrence:")
print(duplicate_rows)
```

```python
# Check for mixed types
for col in df.columns.tolist():
    weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis=1)
    if len (df[weird])>0:
        print(col)
```

```python
#Changing mixed types
df['experience_level'] = df['experience_level'].astype(str)
# Replace or drop non-finite values in 'company_id'
df['company_id'] = pd.to_numeric(df['company_id'], errors='coerce')  # Convert to numeric, replacing non-finite values with NaN
df = df.dropna(subset=['company_id'])  # Drop rows with NaN in 'company_id'

# Convert 'company_id' to integer
df['company_id'] = df['company_id'].astype('int64')
```

```python
# Replcaing nan to Na
df.replace('nan', np.nan, inplace=True)
```

```python
#Finding Missing Values
df.isnull().sum()
```

```python
df= df.dropna(subset=['max_salary'])
```

```python
#Finding Missing Values
df.isnull().sum()
```

```python
df.shape
```

```python
# Create an 'approx_salary' value by taking the mean of the 'min_salary' and 'max_salary'
df['approx_salary'] = df[['min_salary', 'max_salary']].mean(axis=1)
df.head(10)
```

```python
# Drop unnecessary columns
df.drop(columns = ['application_type'],inplace=True)
```

```python
#Export df dataframe 
df.to_csv('job_latestv1.csv', index=False)
```

```python
# Calculate the average views for each industry
average_applies_by_industry = merged_2_df.groupby('skill_abr')['views'].transform('mean')

# Filling missing values with the corresponding industry average
merged_2_df['views'] = merged_2_df['views'].fillna(average_applies_by_industry)
```

```python
# Calculate the average company_size for each industry
average_applies_by_industry = merged_2_df.groupby('skill_abr')['company_size'].transform('mean')

# Filling missing values with the corresponding industry average
merged_2_df['company_size'] = merged_2_df['company_size'].fillna(average_applies_by_industry)
```
