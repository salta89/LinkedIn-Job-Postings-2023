```python
import pandas as pd

# Assuming 'df' is your DataFrame with the 'industry' column

# Define a function to group industries
def group_industry(industry):
    if industry in ['Banking','Financial Services', 'Insurance', 'Investment Banking', 'Venture Capital and Private Equity Principals']:
        return 'Financial Services'
    elif industry in ['Staffing and Recruiting']:
        return 'Staffing and Recruiting'
    elif industry in ['Retail Office Equipment', 'Retail', 'Retail Apparel and Fashion', 'Retail Groceries']:
        return 'Retail'
    elif industry in ['Wholesale', 'Wholesale Building Materials']:
        return 'Wholesale'
    elif industry in ['Hospitals and Health Care', 'Mental Health Care', 'Medical Device', 'Medical Practices']:
        return 'Health Care'
    elif industry in ['Biotechnology Research', 'Biotechnology', 'Pharmaceuticals']:
        return 'Biotechnology'
    elif industry in ['Entertainment', 'Entertainment Providers', 'Performing Arts', 'Gambling Facilities and Casinos',
                      'Musicians', 'Broadcast Media', 'Broadcast Media Production and Distribution', 'Online Audio and Video Media']:
        return 'Entertainment'
    elif industry in ['Food Production', 'Food and Beverage Manufacturing', 'Food and Beverages', 'Beverage Manufacturing',
                      'Dairy Product Manufacturing', 'Wine and Spirits']:
        return 'Food Production'
    elif industry in ['Manufacturing', 'Defense and Space Manufacturing',
                      'Furniture and Home Furnishings Manufacturing', 'Automation Machinery Manufacturing', 'Textile Manufacturing',
                      'Chemical Manufacturing', 'Semiconductor Manufacturing', 'Machinery Manufacturing',
                      'Appliances, Electrical, and Electronics Manufacturing', 'Industrial Machinery Manufacturing',
                      'Packaging and Containers Manufacturing', 'Railroad Manufacture', 'Plastics']:
        return 'Manufacturing'
    elif industry in ['Computer Hardware Manufacturing','Software Development', 'Computer Software', 'Information Technology and Services',
                      'Computer and Network Security', 'Internet', 'Technology, Information and Internet', 'Semiconductors','IT Services and IT Consulting','Computers and Electronics Manufacturing','Computer Software']:
        return 'Technology and IT'
    elif industry in ['Construction', 'Building Materials']:
        return 'Construction'
    elif industry in ['Retail Apparel and Fashion', 'Apparel and Fashion', 'Luxury Goods and Jewelry','Cosmetics', 'Design', 'Design Services']:
        return 'Apparel and Fashion'
    elif industry in ['Hospitality', 'Restaurants', 'Travel Arrangements', 'Events Services']:
        return 'Hospitality'
    elif industry in ['Telecommunications', 'Wireless']:
        return 'Telecommunications'
    elif industry in ['Education Administration Programs', 'Higher Education', 'Primary and Secondary Education', 'E-Learning Providers']:
        return 'Education'
    elif industry in ['Management Consulting', 'Business Consulting and Services', 'Consulting', 'Outsourcing and Offshoring Consulting']:
        return 'Consulting and Business Services'
    elif industry in ['Real Estate']:
        return 'Real Estate'
    elif industry in ['Non-profit Organizations', 'Non-profit Organization Management', 'Civic and Social Organizations', 'Fundraising']:
        return 'Non-profit'
    elif industry in ['Mining', 'Mining and Metals', 'Oil and Gas', 'Oil and Energy']:
        return 'Mining'
    elif industry in ['Government Administration', 'Public Safety', 'Public Policy Offices', 'Public Relations and Communications', 'Public Policy']:
        return 'Government'
    elif industry in ['Legal Services', 'Law Practice', 'Administration of Justice', 'Judiciary']:
        return 'Legal'
    elif industry in ['Automotive', 'Motor Vehicle Manufacturing','Industrial Automation']:
        return 'Automotive'
    elif industry in ['Human Resources Services', 'Human Resources']:
        return 'Human Resources'
    elif industry in ['Renewable Energy']:
        return 'Renewable Energy'
    elif industry in ['Semiconductor Manufacturing']:
        return 'Semiconductor Manufacturing'
    elif industry in ['Defense and Space', 'Aviation and Aerospace', 'Aviation and Aerospace Component Manufacturing', 'Airlines and Aviation', 'Armed Forces']:
        return 'Aerospace and Aviation'
    elif industry in ['Transportation/Trucking/Railroad', 'Logistics and Supply Chain', 'Package/Freight Delivery']:
        return 'Logistics and Transportation'
    elif industry in ['Newspaper Publishing', 'Publishing', 'Media Production', 'Online Media', 'Book and Periodical Publishing']:
        return 'Media and Publishing'
    elif industry in ['Environmental Services', 'Renewables and Environment','Photography','Artists and Writers','Information Services']:
        return 'Environmental and Sustainability'
    else:
        return 'Other'

# Create a new column 'grouped_industry' based on the grouping function
df['grouped_industry'] = df['industry'].apply(group_industry)

# Display the DataFrame with the new 'grouped_industry' column
print(df[['industry', 'grouped_industry']])
```

```python
# Assuming you've already created the 'grouped_industry' column
unique_grouped_industries = df['grouped_industry'].unique()

# Print the unique values
print(unique_grouped_industries)
```

```python
# Assuming 'df' is your DataFrame with the 'grouped_industry' column

# Filter the DataFrame to include only rows where 'grouped_industry' is 'Other'
other_category_df = df[df['grouped_industry'] == 'Other']

# Print unique values in the 'industry' column for the "Other" category
unique_values_other_category = other_category_df['industry'].unique()

# Display the unique values in the "Other" category
print(unique_values_other_category)

```
